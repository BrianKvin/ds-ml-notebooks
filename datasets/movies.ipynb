## Question 1
Who won the Oscar for “Actor in a Leading Role” in 2015?

```python
SELECT *
FROM oscars
WHERE winner = 1.0 AND year = 2015 AND award = "Actor in a Leading Role"
;
```

## Question 2
Write a query that will produce the ten oldest movies in the database?

```python
SELECT *
FROM movies
WHERE release_date IS NOT NULL
ORDER BY release_date ASC
LIMIT 10;
```

## QUestion 3 
How many unique awards are there in the Oscars table?

```python
SELECT COUNT(DISTINCT award)
FROM oscars
; 
```

## Question 4 
How many movies are there that contain the word “Spider” within their title?

## Question 5
How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?

```python
SELECT COUNT(DISTINCT title)
FROM movies AS m
LEFT JOIN genremap AS gm
ON gm.movie_id = m.movie_id
LEFT JOIN genres AS g
ON g.genre_id = gm.genre_id
LEFT JOIN keywordmap AS kwm
ON kwm.movie_id = m.movie_id
LEFT JOIN keywords as kw
ON kw.keyword_id = kwm.keyword_id
 WHERE genre_name ='Thriller' AND keyword_name LIKE '%love%'
;
```

## QUestion 6
How many movies are there that were released between 1 August 2006 ('2006-08-01') and 1 October 2009 ('2009-10-01') that have a popularity score of more than 40 and a budget of less than 50 000 000?

```python
SELECT COUNT(*)
FROM movies 
WHERE release_date BETWEEN ('2006-08-01') AND ('2009-10-01') AND budget < 50000000 AND popularity > 40
;
```

## Question 7
How many unique characters has "Vin Diesel" played so far in the database?

```python

SELECT COUNT(DISTINCT characters)
FROM actors AS a
LEFT JOIN casts AS c
ON c.actor_id = a.actor_id
 WHERE actor_name = 'Vin Diesel';
```

## Question 8
What are the genres of the movie “The Royal Tenenbaums”?

```python

SELECT g.genre_name
FROM movies m
LEFT JOIN genremap gm
ON gm.movie_id =m.movie_id
LEFT JOIN genres g
ON g.genre_id=gm.genre_id
WHERE original_title = 'The Royal Tenenbaums'
```

## Question 9
What are the three production companies that have the highest movie popularity score on average, as recorded within the database?

```python

WITH CTE_topperformers AS (
SELECT
    pc.production_company_id AS pc_id,
    m.movie_id,
    m.title,
    m.release_date,
    m.popularity,
    pc.production_company_name,
    AVG(m.popularity) OVER (PARTITION BY pc.production_company_id) AS avgscore
FROM productioncompanies AS pc
LEFT JOIN productioncompanymap pcm
ON pcm.production_company_id = pc.production_company_id
LEFT JOIN movies m
ON pcm.movie_id = m.movie_id
)
SELECT *
FROM CTE_topperformers
order by avgscore desc
limit 3;
```

## Question 12
Which award category has the highest number of actor nominations (actors can be male or female)?

```python
SELECT award, COUNT(*) AS award_count
FROM (
    SELECT o.award
    FROM oscars o
    WHERE o.name IN (SELECT actor_name FROM actors)
) AS matched_awards
GROUP BY award
ORDER BY award_count DESC
LIMIT 5;
```
